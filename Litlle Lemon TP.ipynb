```python
try:
    # Connect to MySQL
    conn = mysql.connector.connect(**dbconfig)
    cursor = conn.cursor()

    # Create the stored procedure
    create_sp_query = """
    CREATE PROCEDURE peak_hours()
    BEGIN
        SELECT HOUR(BookingSlot) AS HourOfDay, COUNT(*) AS NumBookings
        FROM Bookings
        GROUP BY HourOfDay
        ORDER BY NumBookings DESC;
    END
    """
    # use store procedure
    cursor.execute(create_sp_query)
    print("Stored procedure PeakHours created successfully.")

    # call store procedure
    cursor.callproc('peak_hours')
    # fetch the values in dataset
    results = next(cursor.stored_results())
    dataset = results.fetchall()
    # get the data
    for column_id in cursor.stored_results():
       cols = [column[0] for column in column_id.description]
       print(cols)
    # print the data
    for data in dataset:
       print(data)
    # Commit the transaction
    conn.commit()

except mysql.connector.Error as err:
    print(f"Error: {err}")

finally:
    # Close the cursor and connection
    cursor.close()
    conn.close()
```

```python
#Task 3 create store procedure GuestStatus

try:
    # Connect to MySQL
    conn = mysql.connector.connect(**dbconfig)
    cursor = conn.cursor()

    # Step five: Run the stored procedure query by invoking execute module on the cursor.
    cursor.execute("DROP PROCEDURE IF EXISTS GuestStatus;")
    
    create_sp_query = """
    CREATE PROCEDURE GuestStatus()
    BEGIN
        SELECT CONCAT(b.GuestFirstName, ' ', b.GuestLastName) AS GuestName,
               CASE e.Role
                   WHEN 'Manager' THEN 'Ready to pay'
                   WHEN 'Assistant Manager' THEN 'Ready to pay'
                   WHEN 'Head Chef' THEN 'Ready to serve'
                   WHEN 'Assistant Chef' THEN 'Preparing Order'
                   WHEN 'Head Waiter' THEN 'Order served'
                   ELSE 'Unknown'
               END AS Status
        FROM Bookings  AS b
        LEFT JOIN Employees e ON b.EmployeeID = e.EmployeeID;
    END
    """
    cursor.execute(create_sp_query)
    print("Stored procedure GuestStatus created successfully.")

    # Step six: Invoke callproc to call the stored procedure.
    cursor.callproc("GuestStatus")

    # Step seven: Fetch the results in a variable called dataset.
    results = next(cursor.stored_results())
    dataset = results.fetchall()

    # Step eight: Extract the names of the columns.
    for column_id in cursor.stored_results():
        cols = [column[0] for column in column_id.description]
    print(cols)
    # Step ten: Print the sorted data using for loop.
    print("Sorted data:")
    for row in dataset:
        print(row)

except mysql.connector.Error as err:
    print(f"Error: {err}")

finally:
    # Step eleven: Close the connection to return it back to the pool.
    if 'conn' in locals() and conn.is_connected():
        cursor.close()
        conn.close()
        print("MySQL connection closed.")
```

```python
from mysql.connector.pooling import MySQLConnectionPool, Error


# Bookings data for each guest
guests = [
    {
        "table_number": 8,
        "first_name": "Anees",
        "last_name": "Java",
        "booking_time": "18:00",
        "employee_id": 6
    },
    {
        "table_number": 5,
        "first_name": "Bald",
        "last_name": "Vin",
        "booking_time": "19:00",
        "employee_id": 6
    },
    {
        "table_number": 12,
        "first_name": "Jay",
        "last_name": "Kon",
        "booking_time": "19:30",
        "employee_id": 6
    }
]

try:
    # Loop through each guest and insert their booking
    for guest in guests:
        # Get a connection from the pool
        connection = pool.get_connection()

        if connection.is_connected():
            # Insert statement
            insert_query = """
                INSERT INTO Bookings (TableNo, GuestFirstName, GuestLastName, BookingSlot, EmployeeID)
                VALUES (%s, %s, %s, %s, %s)
            """
            data = (guest["table_number"], guest["first_name"], guest["last_name"], guest["booking_time"], guest["employee_id"])

            # Execute the insert query
            cursor = connection.cursor()
            cursor.execute(insert_query, data)
            connection.commit()
            print(f"Inserted booking for {guest['first_name']} {guest['last_name']} successfully.")

            # No need to close the connection explicitly; closing the cursor will return connection to pool
            cursor.close()

except Error as e:
    print(f"Error inserting booking: {e}")

finally:
    # Don't close the pool explicitly; let the connections manage themselves
    print("All connections returned to the pool.")

```

```python
#Task 3

#Find manager
try:
    # Connect to the database
    connection = pool.get_connection()

    if connection.is_connected():
        # Query to fetch the name and EmployeeID of the Little Lemon manager
        manager_query = """
            SELECT Name, EmployeeID
            FROM Employees
            WHERE Role = 'Manager'
        """

        #establish connection and execute query
        cursor = connection.cursor(dictionary=True)
        cursor.execute(manager_query)

        #fetch data
        manager_info = cursor.fetchone()

        #print data
        if manager_info:
            print(f"The manager of Little Lemon is {manager_info['Name']} (EmployeeID: {manager_info['EmployeeID']})")
        else:
            print("No manager found for Little Lemon.")
        
        cursor.close()

except Error as e:
    print(f"Error retrieving manager information: {e}")
```

```python
#fullname and booking ID of all guests
try:
    # Connect to the database
    connection = pool.get_connection()

    if connection.is_connected():
        # Query to fetch guests waiting to be seated, sorted by BookingSlot
        waiting_guests_query = """
            SELECT CONCAT(GuestFirstName, ' ', GuestLastName) AS FullName, BookingID
            FROM Bookings
            LEFT JOIN Employees 
            ON Employees.EmployeeID = Bookings.EmployeeID
            WHERE Employees.Role = "Receptionist"
            ORDER BY Bookings.BookingSlot DESC;
        """

        #Establish connection and execute query
        cursor = connection.cursor(dictionary=True)
        cursor.execute(waiting_guests_query)

        #fetch query
        waiting_guests = cursor.fetchall()

        #print data
        if waiting_guests:
            print("Guests waiting to be seated:")
            for guest in waiting_guests:
                print(f"{guest['FullName']} (BookingID: {guest['BookingID']})")
        else:
            print("No guests are currently waiting to be seated.")
        
        cursor.close()

except Error as e:
    print(f"Error retrieving waiting guests information: {e}")

finally:
    if connection and connection.is_connected():
        connection.close()
```

```python
#Task 4
#Create procedure
create_salesreport_proc = """
CREATE PROCEDURE Sales_Report()
BEGIN
SELECT SUM(BillAmount) AS Total_sales,
AVG(BillAmount) AS Average_sale,
MIN(BillAmount) AS Min_bill_paid,
MAX(BillAmount) AS Max_bill_paid
FROM Orders;
END
"""
#get connection
connection = pool.get_connection()
#create cursor
cursor = connection.cursor()
#Execute query
cursor.execute(create_salesreport_proc)
#call procedure
cursor.callproc('Sales_Report')
#retrieve result
results = next(cursor.stored_results())
#retrieve data
dataset = results.fetchall()
#print data
for column_id in cursor.stored_results():
    cols = [column[0] for column in column_id.description]
for data in dataset:
    print('\t', cols[0], ':', data[0])
    print('\t', cols[1], ':', data[1])
    print('\t', cols[2], ':', data[2])
    print('\t', cols[3], ':', data[3])
#close
cursor.close()
connection.close()
```
